```python
#Swap

def swap(x=1,y=3):
    z=x
    x=y
    y=z
    print(x,y)
swap(23,45)
```

```python
#Global

def s():
    global v
    v=10
    print(v)
s()
print(v)
```

```python
def number(a,b):
    for i in range(a,b):
        if i%7==0 or i%5!=0:
            print(i)
        i=i+1
number(2000,3200)
```
